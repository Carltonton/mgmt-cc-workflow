```python
# ============================================================================
# ENVIRONMENT SETUP & CENTRALIZED PARAMETERS
# ============================================================================

import random
import statistics

# 📌 ALL PARAMETERS
SEED = 7
random.seed(SEED)

# --- 分析参数 ---
SAMPLE_SIZE = 20

print(f"Seed: {SEED}, Sample size: {SAMPLE_SIZE}")
```

## 计划

- 假设:
- 要扫描的变量:
- 要记录的指标:


```python
# Run experiment using centralized parameters
values = [random.random() for _ in range(SAMPLE_SIZE)]
summary = {
    "count": len(values),
    "mean": statistics.fmean(values),
    "min": min(values),
    "max": max(values),
}
summary
```

## Results

- 关键观察：
- 意外情况或失效模式：
- 决策：继续、调整方向或停止：


```python
# Record findings in a minimal, copy-pasteable structure
result = {
    "seed": SEED,
    "mean": summary["mean"],
    "range": summary["max"] - summary["min"],
}
result
```
